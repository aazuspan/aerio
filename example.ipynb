# Individual photos
Photos can be loaded and processed individually or in collections.

## Loading, previewing, and describing a photo

```python
photo_path = os.path.join("data", "DWW_3FF_117.tif")

# Either photo size, pixel size, or DPI must be provided
photo = Photo(photo_path, photo_size=(224, 224))

# Summarize photo attributes. Attributes will be calculated based on the attribute that was provided.
print(photo)

# Preview the photo
photo.preview(size=(4, 4))
```

## Fiducials

### Locating and preview fiducials

```python
# Calculate the locations of the fiducials
photo.fiducials.locate(size=(80, 120))

# Preview a fiducial to check that it was correctly located
photo.fiducials.bottom.preview()
```

# Photo collections

## Loading a photo collection

```python
# Select a directory that contains multiple photos
photo_dir = os.path.join("data")
# Generate absolute paths to each photo
photo_paths = utils.list_paths(photo_dir, valid_extensions=[".tif"])
# Load all photos in the directory as a Photo Collection
collection = PhotoCollection(photo_paths, photo_size=(224, 224))

# Preview all photos in the collection
collection.preview()
```

## Accessing photos within a collection
Use indexes to select individual photos

```python
collection[0].preview()
```

## Pre-processing a collection

```python
# Crop all photos to the minimum photo dimensions
collection.crop()
# Match histograms for all photos using one of the photos as a reference
collection.match_histograms()
# Locate fiducials for all photos
collection.locate_fiducials(size=(80, 120))
```
